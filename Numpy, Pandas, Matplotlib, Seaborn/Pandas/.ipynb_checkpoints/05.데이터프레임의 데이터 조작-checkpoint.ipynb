```python
ages = [0, 2, 10, 15, 16, 21, 23, 37, 31, 61, 20, 41, 32, 100]
```

```python
# cut         .
# bins     .    NaN .

bins = [1, 15, 25, 35, 60, 99]
labels = ["", "", "", "", ""]
cats = pd.cut(ages, bins, labels=labels)
cats
```

```python
# cut    Categorical  .
#   categories   , codes      .

type(cats)
```

```python
cats.categories
```

```python
cats.codes
```

```python
df4 = pd.DataFrame(ages, columns=["ages"])
df4["age_cat"] = pd.cut(df4.ages, bins, labels=labels)
df4
```

```python
# qcut            .
#     1000  4     250  .

data = np.random.randn(1000)
cats = pd.qcut(data, 4, labels=["Q1", "Q2", "Q3", "Q4"])
cats
```

```python
pd.value_counts(cats)
```

```python
titanic[["age","alive"]]
```

```python
a = titanic[["age","alive"]]
```

```python
def select_rows(a):
    return a.alive == "no"
```

```python
select_rows(a)
```

```python
b = a.loc[select_rows(a), ["age","alive"]]
```

```python
bins = [1, 15, 25, 35, 60, 99]
labels = ["", "", "", "", ""]
cats = pd.cut(b.age, bins, labels=labels)
cats
```

```python
df4 = pd.DataFrame(b.age, columns=["age"])
df4["age_cat"] = pd.cut(b.age, bins, labels=labels)
df4
```

```python
df4["age_cat"].value_counts()
```
