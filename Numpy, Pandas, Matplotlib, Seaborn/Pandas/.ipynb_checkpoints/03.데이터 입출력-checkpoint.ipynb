```python
import pandas as pd
```

```python
%%writefile sample1.csv
c1, c2, c3
1, 1.11, one
2, 2.22, two
3, 3.33, three
```

```python
pd.read_csv('sample1.csv')
```

```python
%%writefile sample2.csv
1, 1.11, one
2, 2.22, two
3, 3.33, three
```

```python
pd.read_csv('sample2.csv', names=['c1', 'c2', 'c3'])
```

```python
# 만약 테이블 내의 특정한 열을 행 인덱스로 지정하고 싶으면 index_col 인수를 사용한다.

pd.read_csv('sample1.csv', index_col='c1')
```

```python
%%writefile sample3.txt
c1        c2        c3        c4
0.179181 -1.538472  1.347553  0.43381
1.024209  0.087307 -1.281997  0.49265
0.417899 -2.002308  0.255245 -1.10515
```

```python
pd.read_table('sample3.txt', sep='\s+')
```

- 만약 자료 파일 중에 건너 뛰어야 할 행이 있으면 skiprows 인수를 사용한다.

```python
%%writefile sample4.txt
파일 제목: sample4.txt
데이터 포맷의 설명:
c1, c2, c3
1, 1.11, one
2, 2.22, two
3, 3.33, three
```

```python
pd.read_csv('sample4.txt', skiprows=[0, 1])
```

```python
pd.read_table('sample4.txt')
```

- 특정한 값을 NaN으로 취급하고 싶으면 na_values 인수에 NaN 값으로 취급할 값을 넣는다.

```python
%%writefile sample5.csv
c1, c2, c3
1, 1.11, one
2, , two
누락, 3.33, three
```

```python
pd.read_csv("sample5.csv")
```

```python
df = pd.bread_csv('sample5.csv', na_values=['누락'])
df
```

```python
df.to_csv('sample6.csv')
```

- 리눅스나 맥에서는 cat 셸 명령으로 파일의 내용을 확인할 수 있다. 윈도우에서는 type 명령을 사용한다.<br>
느낌표(!)는 셸 명령을 사용하기 위한 IPython 매직 명령이다.

```python
!type sample6.csv
```

```python
# 파일을 읽을 때와 마찬가지로 출력할 때도 sep 인수로 구분자를 바꿀 수 있다.
df.to_csv('sample7.txt', sep='|')
```

```python
!type sample7.txt
```

```python
# 또 na_rep 인수로 NaN 표시값을 바꿀 수도 있다.

df.to_csv('sample8.csv', na_rep='누락')
```

```python
!type sample8.csv
```

https://pandas-datareader.readthedocs.io/en/latest/index.html

```python
import datetime
dt_start = datetime.datetime(2015, 1, 1)
dt_end = "2016, 6, 30"
```

https://fred.stlouisfed.org/series/GDP<br>
https://fred.stlouisfed.org/series/CPIAUCSL<br>
https://fred.stlouisfed.org/series/CPILFESL

```python
import pandas_datareader as pdr

gdp = pdr.get_data_fred('GDP', dt_start, dt_end)
gdp.head()
```

```python
# 데이터 코드에 리스트를 넣으면 여러개의 데이터를 동시에 가져온다.

inflation = pdr.get_data_fred(["CPIAUCSL", "CPILFESL"], dt_start, dt_end)
inflation.tail()
```

```python
dt2_start = datetime.datetime(2018, 1, 1)
dt2_end = "2018, 12, 30"

inflation2 = pdr.get_data_fred(["CPIAUCSL", "CPILFESL"], dt2_start, dt2_end)
inflation2.tail()
```
