```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({
    '': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '': ['', '', '', '', '', '', '']
}, columns=['', ''])
df1
```

```python
df2 = pd.DataFrame({
    '': [1001, 1001, 1005, 1006, 1008, 1001],
    '': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['', ''])
df2
```

- merge     df1, df2          .<br>
          inner join  .

```python
pd.merge(df1, df2)
```

```python
# outer join       .
pd.merge(df1, df2, how='outer')
```

```python
# left, right   ,       .

pd.merge(df1, df2, how='left')
```

```python
pd.merge(df1, df2, how='right')
```

```python
#                   .

df1 = pd.DataFrame({
    '': ['setosa', 'setosa', 'virginica', 'virginica'],
    '': [1.4, 1.3, 1.5, 1.3]},
    columns=['', ''])
df1
```

```python
df2 = pd.DataFrame({
    '': ['setosa', 'virginica', 'virginica', 'versicolor'],
    '': [0.4, 0.3, 0.5, 0.3]},
    columns=['', ''])
df2
```

```python
pd.merge(df1, df2)
```

```python
df1 = pd.DataFrame({
    '': ['', '', ''],
    '': ['2018-01-01', '2018-01-02', '2018-01-01'],
    '': ['20000', '30000', '100000']})
df1
```

```python
df2 = pd.DataFrame({
    '': ['', ''],
    '': ['', '']})
df2
```

```python
pd.merge(df1, df2, on='')

#         _x  _y    .
```

```python
#         left_on, right_on     .

df1 = pd.DataFrame({
    '': ['', '', ''],
    '': [1, 2, 3]})
df1
```

```python
df2 = pd.DataFrame({
    '': ['', '', ''],
    '2': [4, 5, 6]})
df2
```

```python
pd.merge(df1, df2, left_on='', right_on="")
```

```python
#        left_index  right_index  True  .

df1 = pd.DataFrame({
    '': ['', '', '', '', ''],
    '': [2000, 2005, 2010, 2000, 2005],
    '': [9853972, 9762546, 9631482, 3655437, 3512547]})
df1
```

```python
df2 = pd.DataFrame(
    np.arange(12).reshape((6, 2)),
    index=[['', '', '', '', '', ''],
           [2000, 2005, 2000, 2005, 2010, 2015]],
    columns=['1', '2'])
df2
```

```python
pd.merge(df1, df2, left_on=['', ''], right_index=True)
```

```python
df1 = pd.DataFrame(
    [[1., 2.], [3., 4.], [5., 6.]],
    index=['a', 'c', 'e'],
    columns=['', ''])
df1
```

```python
df2 = pd.DataFrame(
    [[7., 8.], [9., 10.], [11., 12.], [13, 14]],
    index=['b', 'c', 'd', 'e'],
    columns=['', ''])
df2
```

```python
pd.merge(df1, df2, how='outer', left_index=True, right_index=True)
```

```python
df1.join(df2, how='outer')
```

```python
#       axis=1  .

df1 = pd.DataFrame(
    np.arange(6).reshape(3, 2),
    index=['a', 'b', 'c'],
    columns=['1', '2'])
df1
```

```python
df2 = pd.DataFrame(
    5 + np.arange(4).reshape(2, 2),
    index=['a', 'c'],
    columns=['3', '4'])
df2
```

```python
pd.concat([df1, df2], axis=1)
```
