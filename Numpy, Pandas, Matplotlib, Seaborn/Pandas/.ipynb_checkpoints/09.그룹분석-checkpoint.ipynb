```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(0)
df2 = pd.DataFrame({
    'key1': ['A', 'A', 'B', 'B', 'A'],
    'key2': ['one', 'two', 'one', 'two', 'one'],
    'data1': [1, 2, 3, 4, 5],
    'data2': [10, 20, 30, 40, 50]
})
df2
```

```python
# groupby    A  B    .

groups = df2.groupby(df2.key1)
groups
```

```python
#  GroupBy        groups  .

groups.groups
```

```python
# A B     sum  .

groups.sum()
```

```python
# GroupBy       groupby     .
#    data1    .

df2.data1.groupby(df2.key1).sum()
```

```python
#    GroupBy      data1  .

df2.groupby(df2.key1)["data1"].sum()   # `GroupBy`   data1   
```

```python
df2.groupby(df2.key1).sum()["data1"]  #     data1  
```

```python
df2
```

```python
#    (key1, key2)   data1  .      .

df2.data1.groupby([df2.key1, df2.key2]).sum()
```

```python
#   unstack      .

df2.data1.groupby([df2["key1"], df2["key2"]]).sum().unstack("key2")
```

```python
import seaborn as sns
iris = sns.load_dataset("iris")
```

```python
iris
```

```python
#           .
#          agg  .

def peak_to_peak_ratio(x):
    return x.max() / x.min()

iris.groupby(iris.species).agg(peak_to_peak_ratio)
```

```python
# describe     (descriptive statistics)   .
#          .

iris.groupby(iris.species).describe().T
```

```python
# apply   describe      ( )       . 
#         (petal length)  3    .

def top3_petal_length(df):
    return df.sort_values(by="petal_length", ascending=False)[:3]

iris.groupby(iris.species).apply(top3_petal_length)
```

```python
def q3cut(s):
    return pd.qcut(s, 3, labels=["", "", ""])

iris["petal_length_class"] = iris.groupby(iris.species)["petal_length"].transform(q3cut)
iris[["petal_length", "petal_length_class"]].tail(10)
```

```python
tips = sns.load_dataset("tips")
tips.tail()
```

```python
#             .
#      tip_pct .

tips['tip_pct'] = tips['tip'] / tips['total_bill']
tips.tail()
```

```python
#        .

tips.describe()
```

```python
tips.groupby("sex").count()
```

```python
#    NaN       .
#   size     . size  NaN   .

tips.groupby("sex").size()
```

```python
#       .

tips.groupby(["sex", "smoker"]).size()
```

```python
#          .

tips.pivot_table("tip_pct", "sex", "smoker", aggfunc="count", margins=True)
```

```python
#         .

tips.groupby("sex")[["tip_pct"]].mean()
```

```python
tips.groupby("smoker")[["tip_pct"]].mean()
```

```python
# pivot_table    .

tips.pivot_table("tip_pct", "sex")
```

```python
tips.pivot_table("tip_pct", ["sex", "smoker"])
```

```python
tips.pivot_table("tip_pct", "sex", "smoker")
```

```python
#          .
#      (variance)     describe      .

tips.groupby("sex")[["tip_pct"]].describe()
```

```python
tips.groupby("smoker")[["tip_pct"]].describe()
```

```python
tips.groupby(["sex", "smoker"])[["tip_pct"]].describe()
```

```python
#           .
#             agg  .

def peak_to_peak(x):
    return x.max() - x.min()

tips.groupby(["sex", "smoker"])[["tip"]].agg(peak_to_peak)
```

```python
#          .

tips.groupby(["sex", "smoker"]).agg(["mean", peak_to_peak])[["total_bill"]]
```

```python
#           ( )  .

tips.groupby(["sex", "smoker"]).agg(
    {'tip_pct': 'mean', 'total_bill': peak_to_peak})
```

```python
#  pivot_table      .

tips.pivot_table(['tip_pct', 'size'], ['sex', 'day'], 'smoker')
```

```python
tips.pivot_table('size', ['time', 'sex', 'smoker'], 'day',
                 aggfunc='sum', fill_value=0)
```
