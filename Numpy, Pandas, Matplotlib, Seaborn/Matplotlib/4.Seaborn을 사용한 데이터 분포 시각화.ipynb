```python
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

rugplot : http://seaborn.pydata.org/generated/seaborn.rugplot.html

```python
plt.rcParams['figure.figsize'] = [10,5]
```

```python
x = iris.petal_length.values
```

```python
print(set(x))
```

```python
# matplotlib의 단순한 히스토그램
plt.hist(x)
plt.show()
```

```python
sns.rugplot(x)
plt.title("Iirs 데이터 중, 꽃잎의 길이에 대한 Rug Plot")
plt.show()
```

kdeplot: http://seaborn.pydata.org/generated/seaborn.kdeplot.html
<br> - http://scikit-learn.org/stable/modules/density.html

```python
sns.kdeplot(x)
plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Kernel Density Plot")
plt.show()
```

distplot: http://seaborn.pydata.org/generated/seaborn.distplot.html

```python
sns.distplot(x, kde = True, rug = True)
plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Dist Plot")
plt.show()
```

```python
sns.distplot(x, kde = True, rug = False)
plt.title("Iris 데이터 중, 꽃잎의 길이에 대한 Dist Plot (rug 제외)")
plt.show()
```

jointplot : http://seaborn.pydata.org/generated/seaborn.jointplot.html

```python
sns.jointplot(x = 'sepal_length', y = 'sepal_width', data = iris, kind = 'scatter')
plt.suptitle("꽃받침의 길이와 넓이의 Joint Plot", y=1.05)
plt.show()
```

```python
sns.jointplot(x = 'sepal_length', y = 'sepal_width', data = iris, kind = 'kde')
plt.suptitle("꽃받침의 길이와 넓이의 Joint Plot와 Kernel Density Plot", y=1.05)
plt.show()
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'kde')
plt.show()
```

heatmap: http://seaborn.pydata.org/generated/seaborn.heatmap.html

```python
titanic_size = titanic.pivot_table(index = 'class', columns = 'sex', aggfunc = 'size')
titanic_size
```

```python
sns.heatmap(titanic_size, annot = True, fmt = 'd')
plt.title("Heatmap")
plt.show()
```

```python
sns.heatmap(titanic_size, cmap = sns.light_palette("gray", as_cmap=True), annot = True, fmt = 'd')
plt.title("Heatmap")
plt.show()
```
