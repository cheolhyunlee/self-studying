```python
mpl.rc('font', family='nanumgothic')
mpl.rc('axes', unicode_minus=False)
y = [2,3,1]

x = np.arange(len(y))
xlabel = ['', '', '']
plt.title(" ")
plt.bar(x,y)
plt.xticks(x, xlabel)
plt.yticks(sorted(y))
plt.xlabel("")
plt.ylabel(" ")
plt.show()
```

```python
# xerr  yerr    (error bar)   .
#   alpha  . 0  , 1  .

np.random.seed(0)

people = ['', '', '', '']
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.title("Barh Chart")
plt.barh(y_pos, performance, xerr=error, alpha=0.5)
plt.yticks(y_pos, people)
plt.xlabel('x ')
plt.show()
```

```python
x = np.linspace(0.1, 2 * np.pi, 10)
plt.title("Stem Plot")
plt.stem(x, np.cos(x), '-.')
plt.show()
```

http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.pie

```python
labels = ['', '', '', '']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.title("Pie Chart")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```

http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.contour
http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.contourf

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)

XX, YY = np.meshgrid(x, y)
ZZ = f(XX, YY)
```

```python
plt.title("Contour plots")
plt.contourf(XX, YY, ZZ, alpha = .75, cmap='jet')
plt.contour(XX, YY, ZZ, colors='black')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
XX, YY = np.meshgrid(X, Y)
RR = np.sqrt(XX**2 + YY**2)
ZZ = np.sin(RR)
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.set_title("3D Surface Plot")
ax.plot_surface(XX, YY, ZZ, rstride=1, cstride=1, cmap='hot')
plt.show()
```
