```python
x = np.arange(3)
y = np.arange(5)
print(x)
print(y)
```

```python
X , Y = np.meshgrid(x,y)
```

```python
X
```

```python
Y
```

```python
[list(zip(x, y)) for x, y in zip(X, Y)]
```

```python
import matplotlib.pyplot as plt
plt.title("np.meshgrid  Grid Points")
plt.scatter(X, Y, linewidths=5)
plt.show()
```
