### NumPy 선형방정식,행렬식

```python
import numpy as np
a = np.array([[2,3],[1,-2]])
b = np.array([1,4])
x=np.linalg.solve(a,b)
print(x)
```

```python
import numpy as np
a = np.array([[2,3,1],[1,-2,8],[4,5,9]])
b = np.array([1,4,9])
x=np.linalg.solve(a,b)
print(x)
```

```python
f= a@x
f
```
