## 배열 인덱싱
- 원소의 위치 파악
- 1차원 배열의 경우 인덱스 값이 파이썬 리스트의 인덱스 값과 동일하다.
- 2차원 배열과 3차원 배열의 경우 인덱스를 2개, 3개 사용하여 개별 원소 참조

        배열이름[X]

```python
import numpy as np
```

```python
a = np.array([1,2,3,4])
```

```python
print(a[1])
print(a[3])
print(a[-1])
```

```python
print(a[np.array([0,1])])
print(a[np.array([2,3])])
print(a[np.array([0,1,3])])
```

```python
b = np.arange(0,6).reshape(3,2)
print(b)
```

```python
print(b[0,0])
print(b[1,1])
print(b[0,1])
print(b[2,0])
```

```python
a = np.array([[1,2,3],[4.5,5,6]],int)
print(a)
print(a[0,0])
print(a[0][0])
```

```python
a[0,0] = 12
print(a)
```

```python
a[0][0] = 5
print(a)
```

```python
a = np.arange(0,24).reshape(4,3,2)
```

```python
print(a[1,2,1])
print(a[2,1,1])
print(a[3,2,0])
print(a[0,0,1])
```

```python
print(a[0,0])
print(a[0,1])
print(a[0])
print(a[2,0,1])
print(a[1,1,1])
```
