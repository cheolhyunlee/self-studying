## Numpy 배열의 생성과 reshape, axis

### 배열 생성에 관련된 함수
    zeros((n,m)) 함수 : n x m 배열(행렬)을 생성하며 초기값은 0으로 한다.
    ones((n,m)) 함수 : n x m 배열(행렬)을 생성하며 초기값은 1으로 한다.
    full((n,m),x) 함수 : n x m 배열(행렬)을 생성하며 초기값은 x으로 한다.
    eye(n) 함수 : n x n 크기의 항등행렬(Identity Matrix)를 생성한다.

```python
import numpy as np
```

```python
a = np.zeros((2,3))
a
```

```python
b = np.zeros(4)
b
```

```python
b = np.zeros((4))
b
```

```python
b = np.zeros((1,4))
b
```

```python
c = np.zeros((3,4,5))
c
```

```python
a = np.ones((4,5))
a
```

```python
a = np.full((2,3),55)
a
```

```python
a = np.full((3),4)
a
```

```python
a = np.full(3,4)
a
```

```python
g = np.empty((4, 3))
g
```

```python
np.identity(3, dtype=np.int16)
```

```python
np.identity(5)
```

```python
a = np.eye(3)
a
```

```python
np.eye(3,5)
```

```python
np.eye(3,5,1)
```

```python
a = np.arange(9).reshape(3,3)
a
```

```python
np.diag(a)
```

```python
np.diag(a,0)
```

```python
np.diag(a,1)
```

```python
np.diag(a,2)
```

```python
a = np.arange(0,24).reshape(6,4)
a
```

```python
np.zeros_like(a)
```

### 숫자의 열을 생성하는 함수
    arange(m,n) : m에서 n-1까지 연속적인 정수나 실수를 생성
    arange(m,n,step) : m에서 n-1까지 step만큼 건너뛰며 연속적인 정수나 실수 생성
    arange().reshape(m,n) : arange() 함수가 만든 연속적인 값을 m x n 형태의 행렬로 재구성

```python
a = np.arange(0,10)
a
```

```python
a = np.arange(0,10,2)
a
```

```python
a = np.arange(1.8,5.2)
a
```

```python
a = np.arange(1.8,5.2,1)
a
```

```python
a = np.arange(0.8,10.7,0.1)
a
```

```python
a = np.arange(0.8,10.7,0.0001)
a
```

### axis(축)
- 배열에서 axis란 배열의 인덱스를 기술하는 순서<br>


```python
a = np.arange(0,6)
b = a.reshape(3,2)
b.sum(axis=0)
```

```python
b.sum(axis=1)
```
