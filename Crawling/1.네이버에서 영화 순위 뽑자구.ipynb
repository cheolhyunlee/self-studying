```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get("http://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date=20190506")
c = r.content 
```

```python
soup = BeautifulSoup(c, "html.parser")
print(soup.prettify())
```

```python
all = soup.find("tbody")
print(all)
```

```python
all2 = all.find_all("tr")
print(all2)
```

```python
print(all2[3])
```

```python
all3 = all.find_all("a")
print(all3)
```

```python
for item in all3 :
    print(item)
```

```python
for name in all3 :
    print(name.text)
```

```python
list0 = []
for name in all3 :
    list0.append(name.text)
print(list0)    
```

```python
all4 = all.find_all('img')
print(all4)
```

```python
list_proto = []

for tem in all4 :
    list_proto.append(tem["alt"])
```

```python
print(list_proto)
```

```python
list1 = []
for a in range(len(all4)) :
    if a%2 == 0 :
        list1.append(int(list_proto[a]))
    else :
        pass
print(list1)
```
