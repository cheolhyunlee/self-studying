```python
import requests
from bs4 import BeautifulSoup
r = requests.get('https://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=100')
c = r.content
soup = BeautifulSoup(c, "html.parser")
```

```python
all = soup.find("ul", {"class":"type06_headline"})
print(all)
```

```python
all2 = all.find_all("li")
print(all2[4])
```

```python
for item in all2 :
    title = item.find("dt", {"class":""}) # dt 두개라서 클래스를 공백으로
    print(title)
```

```python
for item in all2 :
    title = item.find("dt", {"class":""}).text
    print(title)
```

```python
for item in all2 :
    title = item.find("dt", {"class":""}).text.replace("\t","").replace("\n","") # 공백이랑 탭 없애기
    modifyTitle = title[2:len(title)+1] # 앞에 공백 한개 더 없애기
    print(modifyTitle)
```

```python
for item in all2 :
    img = item.find("dt",{"class":"photo"})
    img2 = img.find("img")["src"]
    print(img2)
```

```python
for item in all2 :
    try :
        img = item.find("dt",{"class":"photo"})
        img2 = img.find("img")["src"]
        print(img2) # 이미지가 없는 기사가 있어서...
    except :
        print("no image")
```

```python
writing = all.find_all("span", {"class":"writing"})

for w in writing :
    print(w.text)
```

```python
for item in all : 
    print(item)
    
writing = all.find_all("span", {"class":"writing"})
```

```python
for item in all : 
    for item2 in item :
        link = item2.find("dt",{"class" : ""}).find("a")
        print(link['href'])
        print(link.text.replace("\t","").replace("\n","")[2:len(link.text)+1])
```

```python
dl = all.find_all("dl")
for item2 in dl :
    link = item2.find("dt",{"class" : ""}).find("a")
    print(link['href'])
    print(link.text.replace("\t","").replace("\n","")[2:len(link.text)+1])
```

```python
for item in dl :
    try :
        content = item.find("dd")
        print(content.text.replace("\t","").replace("\n","").split("...")[0])
        print(content.find("span",{"class":"writing"}).text)
        print(content.find("span", {"class":"date"}).text)
        print(" ")
    except :
        print("No Content")
```
