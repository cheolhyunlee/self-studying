```python
import requests
from bs4 import BeautifulSoup
r = requests.get('https://finance.naver.com/sise/sise_index_day.nhn?code=KOSPI')
c = r.content
soup = BeautifulSoup(c, "html.parser")
```

```python
all = soup.find_all("tr")
print(all)
```

```python
date = soup.find_all("td", {"class" : "date"})
item_date = []
for item in date :
    item_date.append(item.text)
```

```python
number = soup.find_all("td", {"class" : "number_1"})
item_number = []
for a in range(len(number)) :
    for item in number[a] :
        if a%4 == 0 :
            item_number.append(item)
        else :
            pass
```

```python
for a in range(len(item_date)) :
    print("날짜:%s 종가:%s" % (item_date[a],item_number[a])) 
```
