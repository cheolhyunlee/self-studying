# while

### while문의 기본 구조
- while <조건문>:  
    <수행할 문장1>  
    <수행할 문장2>  
    <수행할 문장3>  
    ...  
<br>
- while문은 조건문이 참인 동안에 while문 아래에 속하는 문장들이 반복해서 수행된다.

```python
#"열 번 찍어 안 넘어 가는 나무 없다"
treehit = 0
while treehit < 10 : # while문의 조건문은 treeHit < 10
    treehit = treehit + 1
    print("나무를 %d번 찍었습니다." % treehit)
    if treehit == 10 :
        print("나무 넘어가버렷")
```

### while문 강제로 빠져나가기
- while문은 조건문이 참인 동안 계속해서 while문 안의 내용을 반복적으로 수행한다.  
하지만 강제로 while문을 빠져나가고 싶을 때가 있다.<br><br>
예를 들어 커피 자판기를 생각해 보자.  
자판기 안에 커피가 충분히 있을 때에는 동전을 넣으면 커피가 나온다.  
그런데 자판기가 제대로 작동하려면 커피가 얼마나 남았는지 항상 검사해야 한다.  
만약 커피가 떨어졌다면 판매를 중단하고 "판매 중지"라는 문구를 사용자에게 보여주어야 한다.<br><br>
이렇게 판매를 강제로 멈추게 하는 것이 바로 break문이다.  

```python
#커피 자판기

coffee = 10
money = 300
while money :
    print("돈을 받았으니 커피를 줄껭")
    coffee = coffee -1
    print("남은 커피는 %d개임" % coffee)
    if coffee == 0:
        print("커피가 다 떨어졌엉 ㅠㅠ 이제 안팔래;;")
        break
```

```python
coffee = 10
while True :
    money = int(input("돈을 넣으세요 : "))
    if money == 300 :
        print("커피를 주겠다.")
        coffee = coffee -1
        print("남은 커피는 %d개당" %coffee)
    elif money > 300 :
        print("거스름돈 %d원을 주고 커피를 드림" %(money -300))
        coffee = coffee -1
        print("남은 커피는 %d개당" %coffee)
    else :
        print("돈도 없는게;;; 가라 좀 ㅡㅡ")
        print("남은 커피는 %d개임" %coffee)
    if coffee == 0 :
        print("커피가 다 팔림! 판매중지!")
        break
```

## while문의 맨 처음으로 돌아가기
- while문 안의 문장을 수행할 때 입력된 조건을 검사해서 조건에 맞지 않으면 while문을 빠져나간다.  
그런데 프로그래밍을 하다 보면 while문을 빠져나가지 않고  
while문의 맨 처음(조건문)으로 다시 돌아가게 만들고 싶은 경우가 생기게 된다.  
이때 사용하는 것이 바로 continue문이다.

```python
#만약 1부터 10까지의 숫자 중에서 홀수만 출력하는 것을 while문을 이용해서 작성한다고 생각해 보자. 어떤 방법이 좋을까?
i = 0
while i <= 4 :
    j = 2*i + 1
    i = i + 1
    print(j)

#내가 만든거!!!
```

```python
#답안

a = 0
while a < 10 :
    a = a + 1
    if a % 2 == 0 : continue #a를 2로 나누었을 때 나머지가 0인 경우 / a가 짝수이면 continue 문장을 수행 
    print(a)

#continue문은 while문의 맨 처음(조건문: a<10)으로 돌아가게 하는 명령어
#a가 짝수이면 print(a)는 수행되지 않는다.
```

## 무한루프
- 이번에는 무한 루프(Loop)에 대해서 알아보자.  
무한 루프란 무한히 반복한다는 의미이다.  
우리가 사용하는 일반적인 프로그램 중에서 무한 루프의 개념을 사용하지 않는 프로그램은 거의 없다.  
그만큼 자주 사용된다는 뜻이다.  
파이썬에서 무한 루프는 while문으로 구현할 수 있다.  
<br>
- 다음은 무한 루프의 기본적인 형태이다.  
        while True:  
            수행할 문장1  
            수행할 문장2
- while문의 조건문이 True이므로 항상 참이 된다.  
따라서 while문 안에 있는 문장들은 무한하게 수행될 것이다.  
<br>
- 다음의 무한 루프 예이다.

    while True:  
     print("Ctrl+C를 눌러야 while문을 빠져나갈 수 있습니다.")  
  
    Ctrl+C를 눌러야 while문을 빠져나갈 수 있습니다.  
    Ctrl+C를 눌러야 while문을 빠져나갈 수 있습니다.  
    Ctrl+C를 눌러야 while문을 빠져나갈 수 있습니다.
<br><br>
- 위의 문장은 영원히 출력된다.   
하지만 이 예처럼 아무 의미 없이 무한 루프를 돌리는 경우는 거의 없을 것이다.   
[Ctrl+C]를 눌러 빠져나가도록 하자.

- [문제1] 3의 배수의 합  
1부터 1000까지의 자연수 중 3의 배수의 합을 구하시오.

```python
a = 1
b = 0
while a < 1000 :
    a = a + 1
    if a % 3 == 0 :
        b = b + a
        print('%d번째 합은 %d입니다.' %(a/3,b))
print('끝~')
#내가 푼거!
        
```

```python
a = 1
b = 0
while a < 1000 :
    a = a + 1
    if a % 3 == 0 :
        b = b + a
    if a == 1000 :
        print('1부터 1000까지의 자연수 중 3의 배수의 합은 %d입니다.' %b)
#내가 푼거! 야호!
```

```python
#답안
result = 0
i = 1
while i <= 1000:
    if i % 3 == 0:  # 3으로 나누어 떨어지는 수는 3의 배수
        result += i
    i += 1

print(result)
```

- [문제2] 50점 이상의 총합  
다음은 A학급 학생의 점수를 나타내는 리스트이다.  
다음 리스트에서 50점 이상의 점수들의 총합을 구하시오.  
A = [20, 55, 67, 82, 45, 33, 90, 87, 100, 25]

```python
A = [20, 55, 67, 82, 45, 33, 90, 87, 100, 25]

i = -1
while True :
    i = i + 1
    if A[i] >= 50 :
        print(A[i])
    if i == (len(A)-1) :
        break
```

```python
A = [20, 55, 67, 82, 45, 33, 90, 87, 100, 25]

i = -1
j = 0

while True :
    i = i + 1
    if A[i] >= 50 :
        j += A[i]
    if i == (len(A)-1) :
        break
        
print("50점 이상의 점수들의 총합은 %d입니다." %j)

#내가 푼거
```

```python
#답안
A = [20, 55, 67, 82, 45, 33, 90, 87, 100, 25]

result = 0
while A:  # A 리스트에 값이 있는 동안
    mark = A.pop()  # A리스트의 가장 마지막 항목을 하나씩 뽑아냄
    if mark >= 50:  # 50점 이상의 점수만 더함
        result += mark

print(result)  # 481 출력
```

[문제3] 별 표시하기1  
while문을 이용하여 아래와 같이 별(*)을 표시하는 프로그램을 작성해 보자.
    *  
    **  
    ***  
    ****

```python
i=0
while i < 4 :
    i += 1
    print('*' * i)
    
#내가 푼거
```

```python
#답안
i = 0
while True:
    i += 1  # while문 수행 시 1씩 증가
    if i > 4: break  # i 값이 4보다 크면 while문을 벗어난다.
    print ('*' * i)  # i 값 개수만큼 *를 출력한다.
    
#while문을 수행할 때마다 i 값을 증가시킨다.
#별 모양을 4번 출력해야 하므로 i 값이 4보다 클 경우 while문을 벗어나도록 한다.
#별 모양을 i 값 만큼 출력하기 위해서 문자열 곱하기 기능을 이용한다.
```
