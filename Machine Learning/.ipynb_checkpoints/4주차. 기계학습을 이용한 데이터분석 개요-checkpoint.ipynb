```python
from sklearn import datasets # 데이터 셋
from sklearn.tree import DecisionTreeClassifier # 의사결정 트리 기반 분류기
```

```python
from sklearn.model_selection import train_test_split # 데이터 셋 분리

# Stratified k fold cross validation을 사용하기 위한 모듈
# Kfold 모듈과는 다르게 각 fold내 데이터의 클래스 비율을 일정하게 유지
from sklearn.model_selection import StratifiedKFold

from sklearn.model_selection import cross_val_score # cross validation의 정확도 측정
```

```python
# 성능평가 관련 모듈
from sklearn.metrics import confusion_matrix # confusion matrix를 추출하기 위한 모듈
from sklearn.metrics import accuracy_score # accuracy를 측정하기 위한 모듈
from sklearn.metrics import classification_report # recall, precision, f-measure를 측정하기 위한 모듈
from sklearn.metrics import roc_auc_score # roc 곡선 아래 넓이를 구하기 위한 모듈
from sklearn.metrics import mean_squared_error # MSE를 구하기 위한 모듈
```

```python
data = datasets.load_breast_cancer()
# 증상 X가 있다면 Y 유방암이 있는지 없는지
X = data.data # 악성 양성 판단에 영향을 주는 속성 데이터
y = data.target # 악성 양성 클래스 데이터
```

```python
# holdout : train, test set 구성 (test size : 20%)
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

```python
clf = DecisionTreeClassifier() # DecisionTreeClassifier 모델을 변수 clf에 할당
clf.fit(X_train, y_train) # Train 데이터(X_train, y_train)와 fit 함수를 사용해 모델 훈련
clf
```

```python
# Predict 함수를 사용해 test 데이터(X_test)에 대한 예측 값을 구하고 변수 y_pred에 저장
y_pred = clf.predict(X_test)
```

## 모델 성능평가

### Holdout

```python
print('confusion_matrix')
print(confusion_matrix(y_test, y_pred))
```

```python
# Y_test와 y_pred 값을 비교해 정확도(accuracy)를 출력
# normalize=False : 올바르게 분류된 데이터 건수 출력
# normalize=True : 올바르게 분류된 데이터의 비율 출력

print('Accuracy')
print(accuracy_score(y_test, y_pred, normalize = True))
print(accuracy_score(y_test, y_pred, normalize = False)) # 36+71
```

```python
print('Classification Report')
print(classification_report(y_test, y_pred))
```

```python
# Roc_auc_score 함수를 사용해 roc 곡선 아래 면적 출력
print('AUC')
print(roc_auc_score(y_test, y_pred))
```

```python
print('mean squared error')
print(mean_squared_error(y_test, y_pred))
```

### K fold cross validation - shuffle

```python
# StratifiedKFold 모듈을 사용해 10 cross validation 모듈을 skf로 생성
# 10 fold를 구성하는 모듈 skf
skf = StratifiedKFold(n_splits = 10)
skf.get_n_splits(X,y)
print(skf)
```

```python
# 10번의 실험을 위한 데이터 셋 구성
for train_index, test_index in skf.split(X,y):
    print("train set : ", train_index)
    print("test set : ", train_index)
```

```python
# DecisionTreeClassifier 모델을 clf에 생성
# Cross_val_score 함수를 사용해 X, y 데이터에 대해 10 fold cross validation 진행한 accuracy 출력
# 10개 accuracy의 평균 출력

clf = DecisionTreeClassifier()
scores = cross_val_score(clf,X,y,cv=skf)
print('K fold cross validation scores')
print(scores)
print("average accuracy")
print(scores.mean())
```

```python
# StratifiedKFold 모듈을 사용해 10 cross validation 모듈을 skf로 생성
# Shuffle을 True로 설정하여 데이터를 섞은 후 데이터 셋 구성하도록함

skf_sh = StratifiedKFold(n_splits = 10, shuffle = True)
skf_sh.get_n_splits(X,y)
print(skf_sh)
```

```python
# 10번의 실험을 위한 데이터 셋 구성을 살펴봄
# 앞 실험과 다르게 데이터 인덱스가 섞여 있음을 확인 할 수 있음

for train_index, test_index in skf_sh.split(X,y):
    print('Train set : ', train_index)
    print('Test set : ', test_index)
```

```python
clf = DecisionTreeClassifier()
scores = cross_val_score(clf,X,y,cv=skf)
print('K fold cross validation scores')
print(scores)
print("average accuracy")
print(scores.mean())
```
